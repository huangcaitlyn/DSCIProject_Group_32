## 2) Questions

Broad question: What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?

Specific question: Can player activity level (e.g., average session length, number of sessions, total playtime) predict whether a player subscribes to the newsletter?

The two datasets provide player information and session behaviour information which can help examine what factors are most predictive of subscribing to the newsletter. The target variable, subscribe, is provided in players.csv which can be linked with other demographic variables and with behavioural engagement data found in sessions.csv to predict subscription class. 

## 3) Exploratory Data Analysis and Visualization

### Wrangling

```r
# Convert start and end times 
sessions <- sessions |> 
mutate(start_time = ymd_hms(start_time, quiet = TRUE), end_time = ymd_hms(end_time, quiet = TRUE))

# Calculate session length in minutes 
session_length <- sessions |> 
mutate(session_length = as.numeric(difftime(end_time, start_time, units = "mins"))) 

# Group session data by player 
session_summary <- session_length |> 
group_by(hashedEmail) |> 
summarise(num_sessions = n(), avg_session_length = mean(session_length, na.rm = TRUE))

# Combine player data with session data 
merged <- left_join(players, session_summary, by = "hashedEmail")

# Calculate mean of quantitative variables in players.csv 
mean_summary_players <- players |> 
summarise(across(where(is.numeric), ~ round(mean(.x, na.rm = TRUE), 2)))

merged 
mean_summary_players
```

### Exploratory Visualizations

#### Number of sessions vs. subscription status

```r
number_of_sessions <- ggplot(merged, aes(x = num_sessions, fill = subscribe)) +
  geom_histogram(alpha = 0.6, bins = 6) +
  labs(title = "Number of Sessions vs Subscription Status",
    x = "Number of Sessions",
    y = "Number of Players",
    fill = "Subscribed") 
number_of_sessions
```

From this comparison of number of sessions and subscription status, it can be seen that the less sessions played correlates with a larger ratio of non-subscribers compared to subscribers. As number of sessions increases, there are more subscribers than non-subscribers. 

### Total Played Hours vs. Subscription Status

```r
histogram <- merged |>
ggplot(aes(x = played_hours, fill = subscribe)) + 
                    geom_histogram(binwidth = 5, boundary = 0, position = "dodge") + 
                    labs(x = "Total Played Hours", y = "Number of Players", fill = "Subscription Status", title = "Total Played Hours and Subscription Status") + 
coord_cartesian(xlim = c(0, 50))

histogram
```

In this distribution, it is seen that there are more players in the lower range of total hours played. There is a higher amount of subscribers compared to non-subscribers within the interval of 0-10 hours played. After 10 hours, there is virtually no non-subscribers, and only subscribers, indicating that total played hours has an indication on subscription status.

### Experience vs. Subscription Status

```r
experience <- ggplot(merged |>
                     mutate(experience = fct_relevel(experience, "Amateur", "Beginner", "Regular", "Pro", "Veteran")), aes(x = experience, fill = subscribe)) +
  geom_bar(position = "fill") +
  labs(title = "Experience Level vs Subscription Status",
    x = "Experience Level",
    y = "Proportion of Players",
    fill = "Subscription Status") +
  theme_minimal() 
experience
```

This comparison shows the correlation between experience level and subscription status. Throughout all experience levels, there are more subscribers than non-subscribers. The "Regular" experience level has the lowest relative non-subscribers. However, as experience level increases, subscription proportions do not consistently change, indicating other possible factors in subscription status. 

# 4) Methods and Plan

To address whether player behaviour and activity can predict newsletter subscription, knn classification would be the most appropriate method. The response variable of subscription status is binary and categorical, which can be inputted into knn using various explanatory variables. This model is able to handle both categorical (subscription) and continuous variables (played hours, time, etc.). 

Knn classification does not make many assumptions about linearity or distribution, making it suitable for data that may be irregular. However, one main assumption is that similar activity behaviours will have similar subscription levels, and that predictors are standardized. It also assumes that predictors should not have an impact on one another, and are therefore not highly correlated. 

Potential limitations include: sensitivity to outliers; sensitivity to imbalance data; sensitivity to many predictors; possibility of simplification to linearity. 

I will compare models of different values of k using cross-validation on the training set. The k value giving the most accurate prediction will be the model selected. 

Before applying the model, the the datasets will be merged through hashedEmail, converted into a tidy format for readability, and converted to the necessary numeric or categorical formats; the quantitative predictors will be standardized; missing values will be removed. The dataset will then be split into 70% training and 30% testing, providing sufficient training and stable testing. This split will undergo 5-fold-cross-validation to find the optimal k value. As such, no validation set is needed. 

The selection for the final model will be based on accuracy, sensitivity, and specificity in the testing data.
